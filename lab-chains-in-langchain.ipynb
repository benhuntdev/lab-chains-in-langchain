## SequentialChain

````python
from langchain_classic.chains import SequentialChain  
````

````python
first_prompt = ChatPromptTemplate.from_template(
    "Translate the following customer review into English.\n"
    "Keep the original meaning and tone. Do not summarize.\n\n"
    "REVIEW:\n{review}\n\n"
    "ENGLISH TRANSLATION:"
)

chain_one = LLMChain(
    llm=llm,
    prompt=first_prompt,
    output_key="translated_review"
)
````

````python
second_prompt = ChatPromptTemplate.from_template(
    "Summarize the following customer review in English.\n"
    "Return:\n"
    "1) Sentiment: Positive / Negative / Mixed\n"
    "2) Summary: 1 sentence (max 25 words)\n"
    "3) Pros: 2 bullet points\n"
    "4) Cons: 2 bullet points\n\n"
    "REVIEW:\n{translated_review}"
)

chain_two = LLMChain(
    llm=llm,
    prompt=second_prompt,
    output_key="review_summary"
)
````

````python
# prompt template 3: translate to english or other language
third_prompt = ChatPromptTemplate.from_template(
    "Identify the language of the following customer review.\n"
    "Return ONLY the language name in English (e.g., English, French, Spanish).\n\n"
    "REVIEW:\n{review}\n\n"
    "Language:"
)

chain_three = LLMChain(
    llm=llm,
    prompt=third_prompt,
    output_key="review_language"
)
````

````python
from langchain_core.prompts import ChatPromptTemplate


fourth_prompt = ChatPromptTemplate.from_template(
    "You are a customer support assistant for an online store.\n\n"
    "Customer review (English):\n{translated_review}\n\n"
    "Internal summary of the review:\n{review_summary}\n\n"
    "Write a helpful follow-up message to the customer.\n"
    "Rules:\n"
    "- 3 to 5 sentences\n"
    "- Acknowledge what they said\n"
    "- If negative: apologize + offer a practical next step\n"
    "- If positive: thank them + invite them back\n"
    "- Do NOT invent order details, refunds, or policies.\n"
    "- End with a friendly question.\n"
)

chain_four = LLMChain(
    llm=llm,
    prompt=fourth_prompt,
    output_key="followup_message"
)
````

````python
# overall_chain: input= Review 
# and output= English_Review,summary, followup_message
overall_chain = SequentialChain(
    chains=[chain_one, chain_two, chain_three, chain_four],
    input_variables=["review"],
    output_variables=["translated_review", "review_summary", "review_language", "followup_message"],
    verbose=True
)

````

````python
review = df.Review[5]
overall_chain(review)
````

### Translate: English > Spanish | **Sequential Chain**

````python
# Repeat Exercise: (English -> Spanish translation pipline)

translate_prompt = ChatPromptTemplate.from_template(
    "Translate the following customer review from English into Spanish.\n"
    "Keep the original meaning and tone. Do not summarize.\n\n"
    "REVIEW (EN):\n{review}\n\n"
    "SPANISH TRANSLATION (ES):"
)
translate_chain = LLMChain(
    llm=llm,
    prompt=translate_prompt,
    output_key="spanish_review",
)

# Summarize (Spanish)
summarize_prompt = ChatPromptTemplate.from_template(
    "Resume la siguiente reseÃ±a de un cliente (en espaÃ±ol).\n"
    "Devuelve exactamente:\n"
    "1) Sentimiento: Positivo / Negativo / Mixto\n"
    "2) Resumen: 1 frase (mÃ¡x 25 palabras)\n"
    "3) Pros: 2 viÃ±etas\n"
    "4) Contras: 2 viÃ±etas\n\n"
    "RESEÃA (ES):\n{spanish_review}"
)
summarize_chain = LLMChain(
    llm=llm,
    prompt=summarize_prompt,
    output_key="review_summary",
)

# Detect language of original input
language_prompt = ChatPromptTemplate.from_template(
    "Identify the language of the following customer review.\n"
    "Return ONLY the language name in English (e.g., English, French, Spanish).\n\n"
    "REVIEW:\n{review}\n\n"
    "Language:"
)
language_chain = LLMChain(
    llm=llm,
    prompt=language_prompt,
    output_key="review_language",
)

# Customer support follow-up (Spanish)
followup_prompt = ChatPromptTemplate.from_template(
    "Eres un asistente de atenciÃ³n al cliente para una tienda online.\n\n"
    "ReseÃ±a del cliente (ES):\n{spanish_review}\n\n"
    "Resumen interno:\n{review_summary}\n\n"
    "Escribe un mensaje de seguimiento al cliente.\n"
    "Reglas:\n"
    "- 3 a 5 frases\n"
    "- Si es negativa: pide disculpas + ofrece un siguiente paso prÃ¡ctico\n"
    "- Si es positiva: agradece + invita a volver\n"
    "- NO inventes detalles de pedido, reembolsos o polÃ­ticas\n"
    "- Termina con una pregunta amable\n"
)
followup_chain = LLMChain(
    llm=llm,
    prompt=followup_prompt,
    output_key="followup_message",
)
````

````python
# overall SequentialChain
overall_chain = SequentialChain(
    chains=[translate_chain, summarize_chain, language_chain, followup_chain],
    input_variables=["review"],
    output_variables=[
        "spanish_review",
        "review_summary",
        "review_language",
        "followup_message",
    ],
    verbose=True,
)
````

````python
# quick test run
test_review_en = (
    "Fast delivery and great quality. The scent is strong but pleasant, "
    "and the packaging was neat."
)

result = overall_chain({"review": test_review_en})

print("\nDetected language:", result["review_language"])
print("\nSpanish review:\n", result["spanish_review"])
print("\nSummary:\n", result["review_summary"])
print("\nFollow-up message:\n", result["followup_message"])
````

## Router Chain

````python
physics_template = """You are a very smart physics professor. \
You are great at answering questions about physics in a concise\
and easy to understand manner. \
When you don't know the answer to a question you admit\
that you don't know.

Here is a question:
{input}"""


math_template = """You are a very good mathematician. \
You are great at answering math questions. \
You are so good because you are able to break down \
hard problems into their component parts, 
answer the component parts, and then put them together\
to answer the broader question.

Here is a question:
{input}"""

history_template = """You are a very good historian. \
You have an excellent knowledge of and understanding of people,\
events and contexts from a range of historical periods. \
You have the ability to think, reflect, debate, discuss and \
evaluate the past. You have a respect for historical evidence\
and the ability to make use of it to support your explanations \
and judgements.

Here is a question:
{input}"""


computerscience_template = """ You are a successful computer scientist.\
You have a passion for creativity, collaboration,\
forward-thinking, confidence, strong problem-solving capabilities,\
understanding of theories and algorithms, and excellent communication \
skills. You are great at answering coding questions. \
You are so good because you know how to solve a problem by \
describing the solution in imperative steps \
that a machine can easily interpret and you know how to \
choose a solution that has a good balance between \
time complexity and space complexity. 

Here is a question:
{input}"""

biology_template = """You are an excellent biologist. \
You have a deep understanding of living organisms, \
from the molecular and cellular level to entire ecosystems. \
You are skilled at observing patterns in nature, analyzing biological data, \
and explaining complex processes like evolution, genetics, physiology, and ecology. \
You can clearly communicate how life functions and adapts, \
and you make connections between different biological concepts \
to answer challenging questions.

Here is a question:
{input}"""
````

````python
prompt_infos = [
    {
        "name": "physics", 
        "description": "Good for answering questions about physics", 
        "prompt_template": physics_template
    },
    {
        "name": "math", 
        "description": "Good for answering math questions", 
        "prompt_template": math_template
    },
    {
        "name": "History", 
        "description": "Good for answering history questions", 
        "prompt_template": history_template
    },
    {
        "name": "computer_science",
        "description": "Good for answering computer science questions", 
        "prompt_template": computerscience_template
    },
    {
        "name": "biology",
        "description": "Good for answering biology questions",
        "prompt_template": biology_template
    }
]
````

````python
from langchain_classic.chains.router import MultiPromptChain
from langchain_classic.chains.router.llm_router import LLMRouterChain, RouterOutputParser
from langchain_core.prompts import PromptTemplate  # Updated path for prompts (core package)
````

````python
llm = ChatOpenAI(temperature=0)
````

````python
destination_chains = {}
for p_info in prompt_infos:
    name = p_info["name"]
    prompt_template = p_info["prompt_template"]
    prompt = ChatPromptTemplate.from_template(template=prompt_template)
    chain = LLMChain(llm=llm, prompt=prompt)
    destination_chains[name] = chain  
    
destinations = [f"{p['name']}: {p['description']}" for p in prompt_infos]
destinations_str = "\n".join(destinations)
````

````python
default_prompt = ChatPromptTemplate.from_template("{input}")
default_chain = LLMChain(llm=llm, prompt=default_prompt)
````

````python
MULTI_PROMPT_ROUTER_TEMPLATE = """Given a raw text input to a \
language model select the model prompt best suited for the input. \
You will be given the names of the available prompts and a \
description of what the prompt is best suited for. \
You may also revise the original input if you think that revising\
it will ultimately lead to a better response from the language model.

<< FORMATTING >>
Return a markdown code snippet with a JSON object formatted to look like:
```json
{{{{
    "destination": string \ name of the prompt to use or "DEFAULT"
    "next_inputs": string \ a potentially modified version of the original input
}}}}
```

REMEMBER: "destination" MUST be one of the candidate prompt \
names specified below OR it can be "DEFAULT" if the input is not\
well suited for any of the candidate prompts.
REMEMBER: "next_inputs" can just be the original input \
if you don't think any modifications are needed.

<< CANDIDATE PROMPTS >>
{destinations}

<< INPUT >>
{{input}}

<< OUTPUT (remember to include the ```json)>>"""
````

````python
router_template = MULTI_PROMPT_ROUTER_TEMPLATE.format(
    destinations=destinations_str
)
router_prompt = PromptTemplate(
    template=router_template,
    input_variables=["input"],
    output_parser=RouterOutputParser(),
)

router_chain = LLMRouterChain.from_llm(llm, router_prompt)
````

````python
chain = MultiPromptChain(router_chain=router_chain, 
                         destination_chains=destination_chains, 
                         default_chain=default_chain, verbose=True
                        )
````

````python
chain.run("What is black body radiation?")
````

````python
chain.run("what is 2 + 2")
````

````python
chain.run("Why does every cell in our body contain DNA?")
````

 ### Wholesale Incense Business | **Router Chain**

````python
# Destination prompt templates (wholesale spiritual + incense)

PRODUCT_CATALOG_TEMPLATE = """You are a B2B wholesale product specialist for a spiritual items & incense wholesaler.
Help the customer choose products and navigate the catalog.

Rules:
- Ask 1-2 clarifying questions if needed (e.g., scent profile, brand preference, budget, quantity).
- Recommend up to 5 relevant product types or categories (not made-up SKUs).
- Keep it practical for retailers (gift shops, wellness stores, metaphysical shops).

Customer question:
{input}
"""

PRICING_MOQ_TEMPLATE = """You are a B2B wholesale pricing assistant for a spiritual items & incense wholesaler.

Rules:
- Focus on: MOQ, case pack sizes, tiered pricing, bulk discounts, and payment expectations.
- If prices/MOQs are not provided, say what information is needed instead of inventing numbers.
- Ask for: country, currency, desired quantity, and whether the customer has a business account.

Customer question:
{input}
"""

SHIPPING_TEMPLATE = """You are a shipping & logistics assistant for a B2B wholesale spiritual items & incense business.

Rules:
- Help with: shipping destinations, lead times, tracking, damaged shipments, and delivery issues.
- Do not invent policies. If policy details are missing, ask what courier/region/order status applies.
- Provide a clear step-by-step next action list.

Customer question:
{input}
"""

RETURNS_CLAIMS_TEMPLATE = """You are a customer support agent handling returns/claims for a B2B wholesale spiritual items & incense business.

Rules:
- Be empathetic and efficient.
- Do not invent refund rules. Ask for order number, photos (if damaged), and item details.
- Provide a short checklist of what the customer should send to support.

Customer question:
{input}
"""

ACCOUNT_VAT_INVOICE_TEMPLATE = """You are a B2B account & invoicing assistant for a wholesale business in the EU.

Rules:
- Help with: account signup, VAT invoice, reverse charge, company details, and payment methods.
- Do not give legal advice. If unsure, suggest checking with an accountant.
- Ask for: buyer country, VAT number (if applicable), and whether they need an invoice pro-forma.

Customer question:
{input}
"""

GENERAL_TEMPLATE = """You are a helpful assistant for a B2B spiritual items & incense wholesaler.

Rules:
- Keep answers concise.
- Ask clarifying questions when needed.
- Do not invent store policies, prices, certifications, or shipping rules.

Customer question:
{input}
"""
````

````python
# prompt registry

PROMPT_INFOS = [
    {
        "name": "product_catalog",
        "description": (
            "Product selection, category navigation, recommendations, "
            "differences between incense types/brands, merchandising advice."
        ),
        "prompt_template": PRODUCT_CATALOG_TEMPLATE,
    },
    {
        "name": "pricing_moq",
        "description": (
            "Wholesale pricing questions, MOQs, case pack quantities, "
            "bulk discounts, and payment expectations."
        ),
        "prompt_template": PRICING_MOQ_TEMPLATE,
    },
    {
        "name": "shipping",
        "description": (
            "Shipping destinations, lead times, tracking, delivery issues, "
            "couriers, damaged parcels."
        ),
        "prompt_template": SHIPPING_TEMPLATE,
    },
    {
        "name": "returns_claims",
        "description": (
            "Returns, refunds, wrong items, damaged items, claims process, "
            "what information support needs."
        ),
        "prompt_template": RETURNS_CLAIMS_TEMPLATE,
    },
    {
        "name": "account_vat_invoice",
        "description": (
            "Business accounts, VAT invoices, reverse charge, EU B2B invoicing, "
            "company details, payment methods."
        ),
        "prompt_template": ACCOUNT_VAT_INVOICE_TEMPLATE,
    },
]
````

````python
# Router prompt + router chain + destination chains

MULTI_PROMPT_ROUTER_TEMPLATE = """Given a raw text input to a language model,
select the model prompt best suited for the input.

You will be given the names of the available prompts and a description of
what the prompt is best suited for. You may also revise the original input
if you think that revising it will ultimately lead to a better response.

<< FORMATTING >>
Return a markdown code snippet with a JSON object formatted to look like:
```json
{{
  "destination": string,  // name of the prompt to use or "DEFAULT"
  "next_inputs": string   // a potentially modified version of the original input
}}

REMEMBER: "destination" MUST be one of the candidate prompt names specified below
OR it can be "DEFAULT" if the input is not well suited for any of the candidates.
REMEMBER: "next_inputs" can be the original input if no modifications are needed.

<< CANDIDATE PROMPTS >>
{destinations}

<< INPUT >>
{input}

<< OUTPUT (remember to include the ```json snippet) >>"""


destinations_str = "\n".join(
f'{info["name"]}: {info["description"]}' for info in PROMPT_INFOS
)


router_prompt = PromptTemplate(
template=MULTI_PROMPT_ROUTER_TEMPLATE,
input_variables=["input"],
partial_variables={"destinations": destinations_str},
output_parser=RouterOutputParser(),
)

router_chain = LLMRouterChain.from_llm(
llm=llm,
prompt=router_prompt,
)
````

````python
# Destination chains + MultiPromptChain (follow notebook pattern)

destination_chains = {
    info["name"]: LLMChain(
        llm=llm,
        prompt=PromptTemplate(
            template=info["prompt_template"],
            input_variables=["input"],
        ),
    )
    for info in PROMPT_INFOS
}

default_chain = LLMChain(
    llm=llm,
    prompt=PromptTemplate(
        template=GENERAL_TEMPLATE,
        input_variables=["input"],
    ),
)

router_bot = MultiPromptChain(
    router_chain=router_chain,
    destination_chains=destination_chains,
    default_chain=default_chain,
    verbose=True,
)
````
